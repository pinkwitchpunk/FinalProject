```python
data = pd.read_csv("C://Users//boswo//HunterStroudWGU//unemployment.csv")
print ("To demonstrate that the CSV file has been properly read in and to show a sample of the data:")
data.head(5)
```

```python
#Several variables that will be used in upcoming visuals
year = data['year'].to_numpy().reshape(-1,1)
unrate = data['unrate'].to_numpy().reshape(-1,1)
men = data['unrate_men'].to_numpy().reshape(-1,1)
women = data['unrate_women'].to_numpy().reshape(-1,1)
age1 = data['unrate_16_to_17'].to_numpy().reshape(-1,1)
age2 = data['unrate_18_to_19'].to_numpy().reshape(-1,1)
age3 = data['unrate_20_to_24'].to_numpy().reshape(-1,1)
age4 = data['unrate_25_to_34'].to_numpy().reshape(-1,1)
age5 = data['unrate_35_to_44'].to_numpy().reshape(-1,1)
age6 = data['unrate_45_to_54'].to_numpy().reshape(-1,1)
age7 = data['unrate_55_over'].to_numpy().reshape(-1,1)
```

```python
#first visualization, helps to show the user what regular unemployment looks like for a baseline understanding
# before looking into the categories described.
plt.plot(year,unrate)
plt.xlabel("Year")
plt.ylabel("Unemployment Percentage")
plt.title("Total Unemployment Rate Trends")
plt.show()
```

```python
#heat map to demonstrate the correalations between the variables.
print("This is a heat map of the data. It shows how closely the values are correlated with one another.")
print("A value of 1 means it is the same dataset, while a value of 0 means there is no correlation at all.")
print("This means that any value with a high decimal has a very high correalation between the two datasets compared.")
sns.heatmap(data.corr(), cmap="YlGnBu", annot = True)
plt.show()
```
